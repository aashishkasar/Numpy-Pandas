###### 7. Write a Pandas program to replace nulls, with median or mean of the specified columns in a given DataFrame.

```python
### code here
data=pd.read_csv("D:\\Innomatics\\Pandas_assignment\\Pandas_day_2\\titanic.csv")

mean_value=data["fare"].mean()                                    #replace nulls with mean
data["fare"].fillna(value=mean_value, inplace=True)
data.head()
```

```python
median_value=data["fare"].median()                          #replace nulls with median 
data["fare"].fillna(value=median_value, inplace=True)
data.head()
```
